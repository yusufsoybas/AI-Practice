```python
from scipy import stats # scientificalpython (scipy) lineer regresyon için kullanmamız gerekiyor. 
                        #Ardından stats'ı import ediyoruz.
```

```python
linear_regression = stats.linregress(x=df.Sales,
                                   y=df.TV)
```

```python
linear_regression.slope #slope = eğim
```

```python
linear_regression.intercept #intercept = kesişim noktası
```

```python
linear_regression.slope *200 + linear_regression.intercept
```

```python
linear_regression.rvalue
```

```python
r_square = linear_regression.rvalue**2 # **=karesi 
```
