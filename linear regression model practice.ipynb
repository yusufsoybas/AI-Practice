```python
%matplotlib inline
from sklearn.datasets import load_diabetes
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
diabetes = load_diabetes()
X = diabetes.data
y = diabetes.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=11)
print(X_train.shape)
print(X_test.shape)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_regression = LinearRegression()
```

```python
linear_regression.fit(X=X_train, y=y_train)
```

```python
linear_regression.coef_
```

```python
linear_regression.intercept_
```

```python
predicted = linear_regression.predict(X_test)
```

```python
expected = y_test
```

```python
for p, e in zip(predicted[::5], expected[::5]):  # check every 5th element
    print(f'predicted: {p:.2f}, expected: {e:.2f}')
```

```python
# lambda kullanarak tahmin fonksiyonu olu≈üturma
predict = lambda x: linear_regression.coef_ * x + linear_regression.intercept_
```

```python
predict(2019)
```

```python
predict(1890)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python

feature_index = 0 
axes = sns.scatterplot(x=diabetes.data[:, feature_index], y=diabetes.target, hue=diabetes.target, palette='winter', legend=False)
axes.set_ylim(min(y) - 10, max(y) + 10)  # y ekseni √∂l√ßeƒüi ayarlama
```

```python
x = np.linspace(min(X[:, feature_index]), max(X[:, feature_index]), 100)
y = linear_regression.coef_[feature_index] * x + linear_regression.intercept_
plt.plot(x, y, color='red')
plt.show()
```
